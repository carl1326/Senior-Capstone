```python
from math import *
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
sp.init_printing(use_unicode=True, use_latex='mathjax')
%matplotlib inline
```

```python
xpos = np.empty (800)
ypos = np.empty (800)
xvel = np.empty (800)
yvel = np.empty (800)
```

```python
B = 0.004
m = 100
delT = .1
xpos[0] = 0
ypos[0] = 0
theta = pi/4
v = 700
yvel[0] = v * sin(theta)
xvel[0] = v * cos(theta)
g = 9.8
count = 0
```

```python
while (ypos[count] >= 0):
    v = (xvel[count]**2 + yvel[count]**2)**0.5
    xpos[count + 1] = xpos[count] + xvel[count]*delT
    ypos[count + 1] = ypos[count] + yvel[count]*delT
    xvel[count + 1] = xvel[count] - (B*v*xvel[count]/m)*delT
    yvel[count + 1] = yvel[count] - g*delT - (B*v*yvel[count]/m)*delT
    count = count + 1
    
```

```python
plt.plot (xpos, ypos)
plt.xlabel ("Position (m)")
plt.ylabel ("Position (m)")
plt.xlim (0, 22000)
plt.ylim (0, 8000)
```
